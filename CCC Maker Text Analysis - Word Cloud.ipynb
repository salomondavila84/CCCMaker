## Text Analysis for CCC Maker Quarterly Reports Year 1 Q1 through Year 1 Q3 using a word cloud algorithim. 

### A combined text string of all entries for  Activites and Success were analyzed with common words removed.
[Ref WordCloud](http://amueller.github.io/word_cloud/index.html)

```python
import pandas as pd
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```python
report_df = pd.read_csv("cccmaker_y1q1-y1q3.csv")
```

```python
report_df.head()
```

```python
report_df.isnull().sum()
```

```python
report_df['Success'].fillna('-', inplace = True)
```

```python
def words(text_array):
    words = ''
    for text in text_array:
        words = words + ' ' + text
    print (len(words))
    return words
    
```

```python
activities_text = report_df['Activities']
```

```python
stopwords = set(STOPWORDS)
```

```python
nsw = ['-', 'will', 'new'] 
for i in nsw:
    stopwords.add(i)
```

```python
 wordcloud = WordCloud(font_path=None, width=600, height=300, margin=2, prefer_horizontal=0.9, \
                       scale=1, min_font_size=2, relative_scaling=1, stopwords=stopwords, background_color="white")
```

```python
wordcloud.generate(words(activities_text))
```

```python
plt.figure( figsize=(20,10) )
plt.title("Word Cloud for Activities at CCC Maker College")
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
success_text = report_df['Success']
```

```python
wordcloud.generate(words(success_text))
```

```python
plt.figure( figsize=(20,10) )
plt.title("Word Cloud for Success at CCC Maker College")
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

### Conclusion:

#### Some of the most frequest words are "student", "makerspace" and "faculty" followed by words in which they are participating. This result does is reasonable considering that reporting are of college events. Deeper understanding of makerspace utilization may come about through surveys targeted to students and faculty in order to uncover how makerspaces are utilized and how they support college activites.
